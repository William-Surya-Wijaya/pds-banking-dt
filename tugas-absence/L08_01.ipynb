```python
#library import
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#dataset load
df = pd.read_csv('https://raw.githubusercontent.com/shrk-sh-ioai/pds-banking-dt/main/absence-dt/Absenteeism_at_work.csv', sep=',', encoding='cp1252')
print(df.columns);
```

```python
#dataframe load
df2 = df[['Reason for absence', 'Month of absence', 'Day of the week', 'Seasons', 'Transportation expense','Distance from Residence to Work','Service time','Age','Work load Average/day ', 'Hit target', 'Disciplinary failure', 'Education', 'Son', 'Social drinker', 'Social smoker', 'Pet', 'Weight', 'Height', 'Body mass index', 'Absenteeism time in hours']]
```

```python
#heatmap view
corr = df2.corr()
sns.set(rc = {'figure.figsize':(50,50)}, font_scale=2)
sns.heatmap(corr, cmap="YlGnBu", annot = True, linewidths=0, annot_kws={"size": 20})
plt.show()
```

```python
kwargs = dict(alpha = 0.5, bins = 100);

for col in df.columns :
    plt.hist(df[col], **kwargs, color='g', label=col);

    plt.gca().set(title='Histogram '+col, ylabel='Frequency');
    plt.xticks(rotation = 90); plt.legend(); plt.show();
```
